# Reading in data with pandas
* What is Pandas?
* How do I read files using Pandas?

## What is Pandas?
A Python library -> set of functions and data structures
pandas is for data analysis

* Reading data stored in CSV files (other file formats can be read as well)
* Slicing and subsetting data in Dataframes (tables!)
* Dealing with missing data
* Reshaping data (long -> wide, wide -> long)
* Inserting and deleting columns from data structures
* Aggregating data using data grouping facilities using the split-apply-combine paradigm
* Joining of datasets (after they have been loaded into Dataframes)

```python
import pandas as pd
```

## Load in data

```python
# Press 'Tab' key to use autocompletion
df = pd.read_csv("SN7577.tab", sep='\t')
```

```python
# If you forget the sep argument
df_without_sep = pd.read_csv("SN7577.tab")
```

```python
# How to read in the first 10 lines
df_first_10 = pd.read_csv("SN7577.tab", nrows=10, sep='\t')
```

```python
df_first_10
```

```python
# How to read a subset of the columns?
df_subset = pd.read_csv("SN7577.tab", usecols=["Q1", "Q2"], sep='\t')
df_subset2 = pd.read_csv("SN7577.tab", usecols=list(range(0,9)), sep='\t')
```
